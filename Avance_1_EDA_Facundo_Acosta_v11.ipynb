## 7. Estrategias de Limpieza y Correcci√≥n

### Enfoque:
Crear una copia del dataset original y aplicar t√©cnicas de imputaci√≥n inteligente basadas en agrupamientos l√≥gicos.

```python
# =============================================================================
# 7.1 PREPARACI√ìN PARA LA LIMPIEZA
# =============================================================================
print("\n7.1 ESTRATEGIAS DE LIMPIEZA")
print("-" * 50)

# Hacer una copia del dataframe original para preservar los datos crudos
df_clean = df.copy()
print("‚úÖ Copia del dataset creada para preservar datos originales")

# Buscar la columna de visitas correcta
columnas_visitas = [col for col in df_clean.columns if 'visita' in col.lower()]

if columnas_visitas:
    columna_visitas = columnas_visitas[0]
    
    # AN√ÅLISIS Y CORRECCI√ìN DE VALORES NEGATIVOS EN VISITAS
    print(f"\nüîç An√°lisis de valores en {columna_visitas}:")
    print(f"Valor m√≠nimo: {df_clean[columna_visitas].min()}")
    print(f"Valor m√°ximo: {df_clean[columna_visitas].max()}")
    print(f"Tipo de datos: {df_clean[columna_visitas].dtype}")
    
    # Identificar y contar valores negativos
    valores_negativos = df_clean[df_clean[columna_visitas] < 0]
    print(f"\n‚ùå N√∫mero de registros con visitas negativas: {len(valores_negativos)}")
    
    if len(valores_negativos) > 0:
        # Visualizaci√≥n de distribuci√≥n de visitas (incluyendo valores negativos)
        plt.figure(figsize=(12, 5))
        
        plt.subplot(1, 2, 1)
        sns.boxplot(x=df_clean[columna_visitas])
        plt.title(f'Distribuci√≥n de {columna_visitas} (con valores negativos)')
        plt.xlabel('Visitas')
        
        plt.subplot(1, 2, 2)
        sns.histplot(df_clean[columna_visitas], bins=30, kde=True)
        plt.title(f'Histograma de {columna_visitas} (con valores negativos)')
        plt.xlabel('Visitas')
        plt.ylabel('Frecuencia')
        
        plt.tight_layout()
        plt.show()
        
        # ESTRATEGIA DE CORRECCI√ìN PARA VALORES NEGATIVOS EN VISITAS
        print(f"\nüîÑ Correcci√≥n de valores negativos en {columna_visitas}:")
        
        # Calcular la media de visitas (excluyendo valores negativos para c√°lculo)
        media_visitas_positivas = df_clean[df_clean[columna_visitas] >= 0][columna_visitas].mean()
        print(f"üìä Media de visitas (solo valores no negativos): {media_visitas_positivas:.2f}")
        
        # CORRECCI√ìN: Convertir la media al mismo tipo de datos de la columna
        if df_clean[columna_visitas].dtype in ['int64', 'int32', 'int16', 'int8']:
            # Si es entero, redondear la media a entero
            valor_reemplazo = int(round(media_visitas_positivas))
            print(f"üîÑ Media redondeada para mantener tipo entero: {valor_reemplazo}")
        else:
            # Si es float, mantener como float
            valor_reemplazo = media_visitas_positivas
        
        # Reemplazar valores negativos con el valor de reemplazo del tipo correcto
        registros_corregidos = len(df_clean[df_clean[columna_visitas] < 0])
        df_clean.loc[df_clean[columna_visitas] < 0, columna_visitas] = valor_reemplazo
        
        print(f"‚úÖ Valores negativos corregidos: {registros_corregidos} registros")
        print(f"üìà Valor m√≠nimo despu√©s de correcci√≥n: {df_clean[columna_visitas].min()}")
        print(f"üîß Tipo de datos despu√©s de correcci√≥n: {df_clean[columna_visitas].dtype}")
        
        # Visualizaci√≥n despu√©s de la correcci√≥n
        plt.figure(figsize=(10, 6))
        sns.histplot(df_clean[columna_visitas], bins=30, kde=True)
        plt.title(f'Distribuci√≥n de {columna_visitas} (despu√©s de correcci√≥n)')
        plt.xlabel('Visitas')
        plt.ylabel('Frecuencia')
        plt.axvline(valor_reemplazo, color='r', linestyle='--', 
                    label=f'Valor de reemplazo: {valor_reemplazo}')
        plt.legend()
        plt.show()
        
    else:
        print("‚úÖ No hay valores negativos en la columna de visitas")
        
else:
    print("‚ÑπÔ∏è No se encontr√≥ columna de visitas para analizar")
```

```python
# =============================================================================
# 7.2 IMPUTACI√ìN DE VALORES NULOS
# =============================================================================
print("\n7.2 IMPUTACI√ìN DE VALORES NULOS")
print("-" * 50)

# a. Manejo de valores nulos en edad (imputar con mediana)
mediana_edad = df_clean['edad'].median()
df_clean['edad'] = df_clean['edad'].fillna(mediana_edad)
print(f"‚úÖ Valores nulos en edad imputados con mediana global: {mediana_edad}")

# b. IMPUTACI√ìN INTELIGENTE: promedio_gasto_comida por estrato y ciudad
print("\nüçΩÔ∏è Imputaci√≥n de promedio_gasto_comida:")

# Calcular medianas por grupo
medianas_gasto = df_clean.groupby(['estrato_socioeconomico', 'ciudad_residencia'])['promedio_gasto_comida'].median().reset_index()
medianas_gasto = medianas_gasto.rename(columns={'promedio_gasto_comida': 'mediana_grupo'})

# Fusionar las medianas al dataframe principal
df_clean = df_clean.merge(medianas_gasto, on=['estrato_socioeconomico', 'ciudad_residencia'], how='left')

# Imputar valores nulos usando la mediana del grupo correspondiente
gastos_nulos_antes = df_clean['promedio_gasto_comida'].isnull().sum()
df_clean['promedio_gasto_comida'] = df_clean['promedio_gasto_comida'].fillna(df_clean['mediana_grupo'])

# Si todav√≠a hay nulos (grupos sin datos), usar mediana global como respaldo
mediana_global_gasto = df_clean['promedio_gasto_comida'].median()
df_clean['promedio_gasto_comida'] = df_clean['promedio_gasto_comida'].fillna(mediana_global_gasto)

gastos_nulos_despues = df_clean['promedio_gasto_comida'].isnull().sum()
print(f"üìä Valores nulos en gasto imputados: {gastos_nulos_antes} ‚Üí {gastos_nulos_despues}")

# Eliminar la columna temporal
df_clean = df_clean.drop('mediana_grupo', axis=1)

# c. IMPUTACI√ìN INTELIGENTE: preferencias_alimenticias por estrato y ciudad
print("\nü•ó Imputaci√≥n de preferencias_alimenticias:")

def moda_grupo(series):
    return series.mode()[0] if not series.mode().empty else pd.NA

modas_preferencias = df_clean.groupby(['estrato_socioeconomico', 'ciudad_residencia'])['preferencias_alimenticias'].apply(moda_grupo).reset_index()
modas_preferencias = modas_preferencias.rename(columns={'preferencias_alimenticias': 'moda_grupo'})

# Fusionar las modas al dataframe principal
df_clean = df_clean.merge(modas_preferencias, on=['estrato_socioeconomico', 'ciudad_residencia'], how='left')

# Imputar valores nulos usando la moda del grupo correspondiente
pref_nulos_antes = df_clean['preferencias_alimenticias'].isnull().sum()
df_clean['preferencias_alimenticias'] = df_clean['preferencias_alimenticias'].fillna(df_clean['moda_grupo'])

# Si todav√≠a hay nulos, usar moda global como respaldo
moda_global_pref = df_clean['preferencias_alimenticias'].mode()[0]
df_clean['preferencias_alimenticias'] = df_clean['preferencias_alimenticias'].fillna(moda_global_pref)

pref_nulos_despues = df_clean['preferencias_alimenticias'].isnull().sum()
print(f"üìä Valores nulos en preferencias imputados: {pref_nulos_antes} ‚Üí {pref_nulos_despues}")

# Eliminar la columna temporal
df_clean = df_clean.drop('moda_grupo', axis=1)

# d. Para tel√©fono y email: usar pd.NA (mejor pr√°ctica)
df_clean['telefono_contacto'] = df_clean['telefono_contacto'].fillna(pd.NA)
df_clean['correo_electronico'] = df_clean['correo_electronico'].fillna(pd.NA)
print("üìû Tel√©fono y email nulos convertidos a pd.NA")

# Verificaci√≥n final
print(f"\nüîç Verificaci√≥n final de valores nulos:")
nulos_restantes = df_clean.isnull().sum()
columnas_con_nulos = nulos_restantes[nulos_restantes > 0]
if len(columnas_con_nulos) > 0:
    print("Columnas que a√∫n tienen valores nulos:")
    for col, nulos in columnas_con_nulos.items():
        print(f"   - {col}: {nulos} valores nulos")
else:
    print("‚úÖ No hay valores nulos restantes en el dataset")
```

# üéØ JUSTIFICACI√ìN: Usamos mediana para outliers porque es m√°s robusta
La mediana no se "contamina" con valores extremos como s√≠ ocurre con la media
Esto garantiza una imputaci√≥n m√°s conservadora y representativa del comportamiento t√≠pico

```python
# =============================================================================
# 7.3 CORRECCI√ìN DE VALORES INCONSISTENTES
# =============================================================================
print("\n7.3 CORRECCI√ìN DE VALORES INCONSISTENTES")
print("-" * 50)

# Correcci√≥n de valores at√≠picos/inconsistentes en edad
print(f"üîç Registros con edad < 0: {len(df_clean[df_clean['edad'] < 0])}")
print(f"üîç Registros con edad > 120: {len(df_clean[df_clean['edad'] > 120])}")

# Corregir valores imposibles estableciendo l√≠mites razonables
df_clean['edad'] = df_clean['edad'].apply(lambda x: mediana_edad if x < 0 or x > 120 else x)
print("‚úÖ Edades fuera de rango [0, 120] corregidas con la mediana")

# CONVERSI√ìN DE EDAD DE FLOAT A INT
print(f"\nüîÑ Conversi√≥n de tipo de dato de edad:")
print(f"Antes: {df_clean['edad'].dtype}")
df_clean['edad'] = df_clean['edad'].astype(int)
print(f"Despu√©s: {df_clean['edad'].dtype}")
print("‚úÖ Edad convertida de float a int")

# Eliminar duplicados
filas_antes = len(df_clean)
df_clean = df_clean.drop_duplicates()
filas_despues = len(df_clean)
print(f"‚úÖ Filas eliminadas por duplicados exactos: {filas_antes - filas_despues}")

# Verificar que no haya IDs duplicados
if df_clean.duplicated(subset=['id_persona']).sum() > 0:
    print("‚ö†Ô∏è ADVERTENCIA: Todav√≠a hay IDs duplicados despu√©s de la limpieza")
else:
    print("‚úÖ No hay IDs duplicados")
```

### Objetivo:
Identificar relaciones entre variables demogr√°ficas y de comportamiento para segmentaci√≥n de clientes.

```python
# =============================================================================
# 11.1 AN√ÅLISIS CRUZADO: G√âNERO VS VARIABLES CLAVE
# =============================================================================
print("\n11.1 AN√ÅLISIS CRUZADO: G√âNERO VS VARIABLES CLAVE")
print("-" * 50)

variables_cruzadas = ['estrato_socioeconomico', 'preferencias_alimenticias', 'consume_licor', 'membresia_premium']

for variable in variables_cruzadas:
    print(f"\nüîç Distribuci√≥n de {variable} por g√©nero:")
    
    # Tabla de contingencia con porcentajes
    tabla_cruzada = pd.crosstab(df_clean['genero'], df_clean[variable], 
                               margins=True, margins_name='Total')
    
    # Agregar porcentajes por fila
    tabla_porcentaje = pd.crosstab(df_clean['genero'], df_clean[variable], 
                                  normalize='index') * 100
    tabla_porcentaje = tabla_porcentaje.round(2)
    
    print("üìä Frecuencias:")
    print(tabla_cruzada)
    print("\nüìà Porcentajes por g√©nero:")
    print(tabla_porcentaje)
    print("-" * 40)
```

```python
# =============================================================================
# 11.2 AN√ÅLISIS POR CIUDAD Y ESTRATO SOCIOECON√ìMICO
# =============================================================================
print("\n11.2 AN√ÅLISIS POR CIUDAD Y ESTRATO SOCIOECON√ìMICO")
print("-" * 50)

# Preferencias alimenticias por ciudad
print("üçΩÔ∏è Preferencias alimenticias por ciudad:")
pref_ciudad = pd.crosstab(df_clean['ciudad_residencia'], 
                         df_clean['preferencias_alimenticias'],
                         margins=True, margins_name='Total')
print(pref_ciudad)
print("\n")

# Consumo de licor por estrato socioecon√≥mico
print("üç∑ Consumo de licor por estrato socioecon√≥mico:")
licor_estrato = pd.crosstab(df_clean['estrato_socioeconomico'], 
                           df_clean['consume_licor'],
                           normalize='index') * 100
licor_estrato = licor_estrato.round(2)
print(licor_estrato)
print("-" * 40)
```
